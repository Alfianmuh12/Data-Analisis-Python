## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana kepuasan pelanggan berhubungan dengan berbagai variabel (waktu pengiriman, kualitas produk, dll.)?

```python
# Merge tables
merged_data = orders_ollist.merge(order_reviews_ollist, on='order_id', how='inner')
merged_data = merged_data.merge(order_items_ollist, on='order_id', how='inner')
merged_data = merged_data.merge(products_ollist, on='product_id', how='inner')

# Select relevant columns for customer satisfaction analysis
satisfaction_analysis = merged_data[['order_estimated_delivery_date', 'review_score', 'product_category_name']]

# Display the first few rows of the analysis
print(satisfaction_analysis.head())

# Analysis: Average Review Score by Product Category
average_review_by_category = satisfaction_analysis.groupby('product_category_name')['review_score'].mean().reset_index()
average_review_by_category.columns = ['product_category_name', 'average_review_score']

# Display the average review scores
print(average_review_by_category)
```

```python
# Menghitung waktu penerimaan paket dalam hari
orders_ollist['delivery_time'] = (orders_ollist['order_delivered_customer_date'] -
                                   orders_ollist['order_purchase_timestamp']).dt.days

# Menggabungkan dengan tabel ulasan
merged_data = orders_ollist.merge(order_reviews_ollist, on='order_id', how='inner')

# Visualisasi hubungan antara waktu penerimaan paket dan skor ulasan
plt.figure(figsize=(12, 6))
sns.scatterplot(data=merged_data, x='delivery_time', y='review_score', alpha=0.6)

plt.title('Relationship Between Delivery Time and Review Score')
plt.xlabel('Delivery Time (days)')
plt.ylabel('Review Score')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Menggabungkan data untuk mendapatkan informasi produk
merged_data = order_reviews_ollist.merge(orders_ollist, on='order_id', how='inner')
merged_data = merged_data.merge(order_items_ollist, on='order_id', how='inner')
merged_data = merged_data.merge(products_ollist, on='product_id', how='inner')

# Hitung jumlah item terjual dan rata-rata skor ulasan per produk
product_analysis = merged_data.groupby('product_category_name')['order_item_id'].count().reset_index()
product_analysis.columns = ['product_category_name', 'total_items_sold']

average_scores = merged_data.groupby('product_category_name')['review_score'].mean().reset_index()
average_scores.columns = ['product_category_name', 'average_review_score']

# Gabungkan hasil analisis
product_analysis = product_analysis.merge(average_scores, on='product_category_name', how='left')

# Visualisasi
plt.figure(figsize=(12, 6))
sns.barplot(data=product_analysis, x='product_category_name', y='average_review_score')

plt.title('Average Review Score by Product')
plt.xlabel('Product Category Name')
plt.ylabel('Average Review Score')
plt.xticks(rotation=45, ha='right')

# Tambahkan jumlah item terjual di atas batang
for index, row in product_analysis.iterrows():
    plt.text(index, row['average_review_score'], row['total_items_sold'], color='black', ha='center')

plt.tight_layout()
plt.show()
```

Waktu pengiriman yang lebih cepat cenderung meningkatkan kepuasan pelanggan. Pelanggan yang menerima produk lebih cepat sering memberikan skor kepuasan yang lebih tinggi. Ini menunjukkan bahwa kecepatan pengiriman adalah faktor penting dalam pengalaman pelanggan.
Waktu pengiriman yang lama mungkin menyebabkan ketidakpuasan, karena pelanggan mungkin merasa bahwa mereka harus menunggu terlalu lama untuk menerima produk yang mereka pesan.

Kualitas produk memiliki korelasi yang positif signifikan dengan kepuasan pelanggan. Pelanggan yang menerima produk dengan kualitas tinggi cenderung memberikan skor lebih tinggi dalam ulasan mereka.

### Pertanyaan 2: Bagaimana karakteristik pelanggan berdasarkan lokasi (misalnya, kota atau negara bagian)? Apakah ada perbedaan signifikan dalam preferensi produk?

```python
# Menggabungkan data
merged_data = orders_ollist.merge(customers_ollist, on='customer_id', how='inner')
merged_data = merged_data.merge(order_items_ollist, on='order_id', how='inner')
merged_data = merged_data.merge(products_ollist, on='product_id', how='inner')

# Menghitung jumlah pesanan per negara bagian
state_analysis = merged_data.groupby('customer_state')['order_id'].count().reset_index()
state_analysis.columns = ['customer_state', 'total_orders']
```

```python
# Menggabungkan data
merged_data = (orders_ollist
                .merge(customers_ollist, on='customer_id', how='inner')
                .merge(order_items_ollist, on='order_id', how='inner')
                .merge(products_ollist, on='product_id', how='inner'))

# Menghitung jumlah pesanan per kategori produk per negara bagian
category_analysis = (merged_data.groupby(['customer_state', 'product_category_name'])['order_id']
                     .count().reset_index())
category_analysis.columns = ['customer_state', 'product_category_name', 'total_orders']

# Ambil kategori terlaris (jumlah tertinggi) per negara bagian
top_category_per_state = category_analysis.loc[category_analysis.groupby('customer_state')['total_orders'].idxmax()]

# Visualisasi: Kategori Produk Terlaris per Negara Bagian
plt.figure(figsize=(14, 8))
sns.barplot(data=top_category_per_state,
            x='customer_state',
            y='total_orders',
            hue='product_category_name',
            palette='Set2',
            ci=None)

# Judul dan label
plt.title('Top Product Categories by Customer State', fontsize=16)
plt.xlabel('Customer State', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)

# Menyesuaikan sumbu x
plt.xticks(rotation=45, ha='right', fontsize=12)

# Menampilkan legenda dengan format yang lebih jelas
plt.legend(title='Product Category', fontsize=12, title_fontsize='13')

# Menambahkan anotasi jumlah pesanan di atas batang
for p in plt.gca().patches:
    plt.annotate(f'{int(p.get_height())}',
                 (p.get_x() + p.get_width() / 2.,
                  p.get_height()),
                 ha='center',
                 va='bottom',
                 fontsize=10,
                 color='black')

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

## Analisis Lanjutan


**RFM Analysis**

```python
# Load the datasets (adjust file names as per the dataset structure)
orders = pd.read_csv('/content/E-Commerce_dataset/orders_dataset.csv')
order_items = pd.read_csv('/content/E-Commerce_dataset/order_items_dataset.csv')

# Merge orders with items data to get the total monetary value for each order
merged = pd.merge(orders, order_items, on='order_id')

# Calculate 'Monetary' value - total value of each order (sum of order items)
merged['total_price'] = merged['price'] * merged['freight_value']

# Convert 'order_purchase_timestamp' to datetime
merged['order_purchase_timestamp'] = pd.to_datetime(merged['order_purchase_timestamp'])

# Let's create the current date for recency calculation (typically the last date of the dataset)
snapshot_date = merged['order_purchase_timestamp'].max() + pd.DateOffset(days=1)

# Calculate Recency: Time since the last order per customer
recency_df = merged.groupby('customer_id')['order_purchase_timestamp'].max()
recency_df = recency_df.reset_index()
recency_df['Recency'] = (snapshot_date - recency_df['order_purchase_timestamp']).dt.days

# Frequency: Total number of orders per customer
frequency_df = merged.groupby('customer_id')['order_id'].nunique()
frequency_df = frequency_df.reset_index()
frequency_df.rename(columns={'order_id': 'Frequency'}, inplace=True)

# Monetary: Total monetary value spent per customer
monetary_df = merged.groupby('customer_id')['total_price'].sum()
monetary_df = monetary_df.reset_index()
monetary_df.rename(columns={'total_price': 'Monetary'}, inplace=True)

# Combine Recency, Frequency, and Monetary into one DataFrame
rfm_df = pd.merge(recency_df, frequency_df, on='customer_id')
rfm_df = pd.merge(rfm_df, monetary_df, on='customer_id')

# Now let's check the number of unique values in the Frequency column
unique_frequencies = rfm_df['Frequency'].nunique()
print(f"Unique Frequencies: {unique_frequencies}")

# If the number of unique Frequency values is less than 5, adjust the quantiles dynamically
num_quantiles = min(unique_frequencies, 5)

# Apply qcut with the dynamic number of quantiles for Recency, Frequency, and Monetary
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'], num_quantiles, labels=range(1, num_quantiles+1))
rfm_df['Monetary_Score'] = pd.qcut(rfm_df['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# Calculate the RFM score by concatenating the individual scores
rfm_df['RFM_Score'] = rfm_df['Recency_Score'].astype(str) + rfm_df['Frequency_Score'].astype(str) + rfm_df['Monetary_Score'].astype(str)

# Create customer segments based on the RFM score
def rfm_segment(row):
    if row['RFM_Score'] == '555':
        return 'High Value'
    elif row['Recency_Score'] in ['1', '2'] and row['Frequency_Score'] in ['4', '5']:
        return 'Loyal'
    elif row['Recency_Score'] in ['4', '5'] and row['Monetary_Score'] in ['4', '5']:
        return 'At-Risk'
    elif row['Recency_Score'] in ['4', '5'] and row['Frequency_Score'] in ['1', '2']:
        return 'New Customer'
    else:
        return 'Other'

rfm_df['Segment'] = rfm_df.apply(rfm_segment, axis=1)

# Display the final RFM table
print(rfm_df[['customer_id', 'RFM_Score', 'Segment']].head())

# Visualize the distribution of Recency, Frequency, and Monetary
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.histplot(rfm_df['Recency'], bins=20, ax=axes[0], kde=True)
axes[0].set_title('Recency Distribution')

sns.histplot(rfm_df['Frequency'], bins=20, ax=axes[1], kde=True)
axes[1].set_title('Frequency Distribution')

sns.histplot(rfm_df['Monetary'], bins=20, ax=axes[2], kde=True)
axes[2].set_title('Monetary Distribution')

plt.tight_layout()
plt.show()
```
